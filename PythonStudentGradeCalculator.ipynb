```python
#### A simple program to return the grades of a student given the scores
#Capture relevant input
name = input("Enter student name: ")
maths_score = int(input("Enter maths score:"))
physics_score = int(input("Enter physics score:"))
chemistry_score = int(input("Enter chemistry score:"))
biology_score = int(input("Enter biology score:"))

#Output the student name on screen
print(" ")
print(".....",name,"'s performance summary.....")
print(" ")

#the logic for maths
if maths_score>100:
    print("Invalid score for Mathematics. Score cannot be above 100")
elif maths_score>=80:
    print("Mathematics =",maths_score,"grade = A")
elif maths_score>=70:
    print("Mathematics = ",maths_score,"grade = B")
elif maths_score>=60:
    print("Mathematics = ",maths_score,"grade = C")
elif maths_score>=50:
    print("Mathematics = ",maths_score,"grade = D")
elif maths_score<50:
    print("Mathematics = ",maths_score,"grade = E")
else:
    print("Invalid score for Mathematics. Score cannot be below 0")

    #the logic for physics
if physics_score>100:
    print("Invalid score for Physics. Score cannot be above 100")
elif physics_score>=80:
    print("Physics =",physics_score,"grade = A")
elif physics_score>=70:
    print("Physics = ",physics_score,"grade = B")
elif physics_score>=60:
    print("Physics = ",physics_score,"grade = C")
elif physics_score>=50:
    print("Physics = ",maths_score,"grade = D")
elif physics_score<50:
    print("Physics = ",physics_score,"grade = E")
else:
    print("Invalid score for Physics. Score cannot be below 0")
    
    #the logic for biology
if biology_score>100:
    print("Invalid score for Biology. Score cannot be above 100")
elif biology_score>=80:
    print("Biology =",biology_score,"grade = A")
elif biology_score>=70:
    print("Biology = ",biology_score,"grade = B")
elif biology_score>=60:
    print("Biology = ",biology_score,"grade = C")
elif biology_score>=50:
    print("Biology = ",biology_score,"grade = D")
elif biology_score<50:
    print("Biology = ",biology_score,"grade = E")
else:
    print("Invalid score for Biology. Score cannot be below 0")
    
    #the logic for chemistry
if chemistry_score>100:
    print("Invalid score for Chemistry. Score cannot be above 100")
elif chemistry_score>=80:
    print("Chemistry =",chemistry_score,"grade = A")
elif chemistry_score>=70:
    print("Chemistry = ",chemistry_score,"grade = B")
elif chemistry_score>=60:
    print("Chemistry = ",chemistry_score,"grade = C")
elif chemistry_score>=50:
    print("Chemistry = ",chemistry_score,"grade = D")
elif chemistry_score<50:
    print("Chemistry = ",chemistry_score,"grade = E")
else:
    print("Invalid score for Chemistry. Score cannot be below 0")

        #the logic for average score
average_score =(maths_score+physics_score+chemistry_score+biology_score)/4

if average_score>=80:
    print("Average Score =",average_score,"Average grade = A")
    print("PASS")
elif average_score>=70:
    print("Average Score = ",average_score,"Average grade = B")
    print("PASS")
elif average_score>=60:
    print("Average Score = ",average_score,"Average grade = C")
    print("PASS")
elif average_score>=50:
    print("Average Score = ",average_score,"Average grade = D")
    print("PASS")
else:
    print("Average Score = ",average_score,"Average grade = E")
    print("FAIL")
```
